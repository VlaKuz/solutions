```python
import cv2 as cv
import os
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(10, 10))

imgs = []
for i in range(1, 5):
    imgPath = os.path.join('detective', 'image_%d.png' % i)
    imgs.append(cv.imread(imgPath))
    fig.add_subplot(2, 2, i)
    plt.imshow(imgs[-1])
```

```python
for i in range(len(imgs)):
    for j in range(len(imgs)):
        plt.figure()
        plt.imshow(cv.cvtColor(imgs[i] ^ imgs[j], cv.COLOR_BGR2RGB))
```

#### Итоговый ответ (3 картинки ключей):

```python
plt.imshow(cv.cvtColor(imgs[0] ^ imgs[3], cv.COLOR_BGR2RGB))
```

```python
plt.imshow(cv.cvtColor(imgs[1] ^ imgs[3], cv.COLOR_BGR2RGB))
```

```python
plt.imshow(cv.cvtColor(imgs[2] ^ imgs[3], cv.COLOR_BGR2RGB))
```

```python
first_key = imgs[0] ^ imgs[3]
second_key = imgs[1] ^ imgs[3]
third_key = imgs[2] ^ imgs[3]
cv.imwrite('1st_key.png', first_key)
cv.imwrite('2nd_key.png', second_key)
cv.imwrite('3rd_key.png', third_key)
```
