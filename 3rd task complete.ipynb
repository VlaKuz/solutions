```python
import os
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
num_rows = 8
num_cols = 10
parts = []
for y in range(num_rows):
    for x in range(num_cols):
        imgPath = os.path.join('treasure_map', 'row_%d_part_%d.png'% (y, x))
        img = cv.imread(imgPath)
        img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
        parts.append(img)
plt.figure()
plt.imshow(parts[0])
plt.figure()
plt.imshow(parts[1])
plt.figure()
plt.imshow(parts[2])
```

```python
stitcher = cv.createStitcherScans()
```

```python
status, result = stitcher.stitch(parts)
```

```python
plt.imshow(result)
```

```python
cv.imwrite('final_map.png', cv.cvtColor(result, cv.COLOR_RGB2BGR))
```
